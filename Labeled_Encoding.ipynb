```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('/breast_cancer_data.csv')
df.head()
```

```python
df['diagnosis'].value_counts()
```

```python
label = LabelEncoder()
```

```python
#labeling the data with a numeric value
df_label = label.fit_transform(df.diagnosis)
#labels created
df_label
#upending/adding the labels as a seperate column in the dataset
df['labels'] = df_label
df.head()
df['labels'].value_counts()
```
